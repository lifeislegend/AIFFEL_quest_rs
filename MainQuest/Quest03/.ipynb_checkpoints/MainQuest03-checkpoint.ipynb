## Import Libraries

```python
import os, re
import random, math
import numpy as np
import tensorflow as tf

#  history 
import joblib

#     
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model


import warnings 
warnings.filterwarnings(action='ignore')
```

## Set Up Data Load & Directory 

```python
#         
AUTOTUNE = tf.data.experimental.AUTOTUNE
# X-RAY   
IMAGE_SIZE = [180, 180]
res = 180

#  ROOT_PATH
ROOT_PATH = os.path.join(os.getenv('HOME'), 'aiffel/chest_xray/chest_xray')

# train/val/test   
TRAIN_PATH = ROOT_PATH + '/train/*/*.jpeg'
VAL_PATH = ROOT_PATH + '/val/*/*.jpeg'  # val  
TEST_PATH = ROOT_PATH + '/test/*/*.jpeg'


#       
BATCH_SIZE = 16
EPOCHS = 25

print(ROOT_PATH)
```

## Data Load

```python
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)

print(len(train_filenames))
print(len(test_filenames))
print(len(val_filenames))
```

```python
# train  validation   filenames 
filenames = tf.io.gfile.glob(TRAIN_PATH)
filenames.extend(tf.io.gfile.glob(VAL_PATH))

#  filenames 8:2 
train_size = math.floor(len(filenames)*0.8)
random.seed(8)
random.shuffle(filenames)
train_filenames = filenames[:train_size]
val_filenames = filenames[train_size:]

print(len(train_filenames))
print(len(val_filenames))
```

### Labeling Normal and Pneumonia

```python
#      
#     
print(f'Normal image path\n{filenames[0]}')
print(f'Pneumonia image path\n{filenames[2000]}')
```

```python
#    NORMAL       .
COUNT_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
print(f"Normal images count in training set: {COUNT_NORMAL}")

#   PNEUMONIA       .
COUNT_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])
print(f"Pneumonia images count in training set: {COUNT_PNEUMONIA}")
```

```python
train_list_ds = tf.data.Dataset.from_tensor_slices(train_filenames)
val_list_ds = tf.data.Dataset.from_tensor_slices(val_filenames)
```

```python
TRAIN_IMG_COUNT = tf.data.experimental.cardinality(train_list_ds).numpy()
print(f"Training images count: {TRAIN_IMG_COUNT}")

VAL_IMG_COUNT = tf.data.experimental.cardinality(val_list_ds).numpy()
print(f"Validating images count: {VAL_IMG_COUNT}")
```

```python
#           
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == "PNEUMONIA"   #  (True),  (False)
```

```python
#    .
def decode_img(img): #      
    img = tf.image.decode_jpeg(img, channels=3) #  uint8 tensor 
    img = tf.image.convert_image_dtype(img, tf.float32) # float32  
    img = tf.image.resize(img, IMAGE_SIZE) #   IMAGE_SIZE 
    return img

#       .
def process_path(file_path):
    label = get_label(file_path) #  
    img = tf.io.read_file(file_path) #  
    img = decode_img(img) #    
    return img, label
```

```python
# make train, validation dataset

train_ds = train_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
val_ds = val_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
```

```python
# check about image resize, labeling

for image, label in train_ds.take(1): # train_ds.take(1)    .
    print("Image shape: ", image.numpy().shape)
    print("Label: ", label.numpy())
```

```python
# make test dataset

test_list_ds = tf.data.Dataset.list_files(TEST_PATH)
TEST_IMAGE_COUNT = tf.data.experimental.cardinality(test_list_ds).numpy()
test_ds = test_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE)

print(TEST_IMAGE_COUNT)
```

```python
def prepare_for_training(ds, shuffle_buffer_size=1000):
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

train_ds = prepare_for_training(train_ds)
val_ds = prepare_for_training(val_ds)
```

## Data Visualization

```python
#      .
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(BATCH_SIZE):
        ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title("PNEUMONIA")
        else:
            plt.title("NORMAL")
        plt.axis("off")


image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
```

## Build CNN Model

```python
# making conv_block
def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    
    return block

# making Dense Block
def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    
    return block
```

```python
def build_model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
        
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPool2D(),
        
        conv_block(32),
        conv_block(64),
        
        conv_block(128),
        tf.keras.layers.Dropout(0.2),
        
        conv_block(256),
        tf.keras.layers.Dropout(0.2),
        
        tf.keras.layers.Flatten(),
        dense_block(512, 0.7),
        dense_block(128, 0.5),
        dense_block(64, 0.3),
        
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    return model
```

### Data Imbalance Post-Processing

```python
weight_for_0 = (1 / COUNT_NORMAL)*(TRAIN_IMG_COUNT)/2.0 
weight_for_1 = (1 / COUNT_PNEUMONIA)*(TRAIN_IMG_COUNT)/2.0

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
```

## Training Model

1. optimizer: adam
2. loss: binary_crossentropy
3. metrics: accuracy, precision, and recall

```python
with tf.device('/GPU:0'):
    model = build_model()

    METRICS = [
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=METRICS
    )
```

```python
#  fit.    ,     

with tf.device('/GPU:0'):
    history = model.fit(
        train_ds,
        steps_per_epoch=TRAIN_IMG_COUNT // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=val_ds,
        validation_steps=VAL_IMG_COUNT // BATCH_SIZE,
        class_weight=class_weight,
        
    )
```

```python
#baseline
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
```

```python
import os, re
import random, math
import numpy as np
import tensorflow as tf

#  history 
import joblib

#     
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
from sklearn.metrics import confusion_matrix, f1_score, roc_auc_score, accuracy_score, precision_score, recall_score


import warnings
warnings.filterwarnings(action='ignore')

#  ROOT_PATH 
ROOT_PATH = os.path.join(os.getenv('HOME'), 'aiffel/chest_xray/chest_xray')

#    
TRAIN_PATH = ROOT_PATH + '/train/*/*.jpeg'
VAL_PATH = ROOT_PATH + '/val/*/*.jpeg'
TEST_PATH = ROOT_PATH + '/test/*/*.jpeg'

# X-RAY    (Data Augmentation    )
#  Data Augmentation   .
IMAGE_SIZE = [224, 224] # Data Augmentation    
res = IMAGE_SIZE[0] # res  IMAGE_SIZE  

#         
AUTOTUNE = tf.data.experimental.AUTOTUNE

#       
BATCH_SIZE = 16
EPOCHS = 25

print(ROOT_PATH)


#     (  )
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)

filenames = tf.io.gfile.glob(TRAIN_PATH)
filenames.extend(tf.io.gfile.glob(VAL_PATH))

train_size = math.floor(len(filenames)*0.8)
random.seed(8)
random.shuffle(filenames)
train_filenames = filenames[:train_size]
val_filenames = filenames[train_size:]

#       (  )
COUNT_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
COUNT_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])

#    (  )
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == "PNEUMONIA"

def decode_img(img):
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize(img, IMAGE_SIZE)
    return img

def process_path(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, label

# conv_block, dense_block, build_model  (  )
def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    return block

def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    return block

def build_model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),

        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPool2D(),

        conv_block(32),
        conv_block(64),

        conv_block(128),
        tf.keras.layers.Dropout(0.2),

        conv_block(256),
        tf.keras.layers.Dropout(0.2),

        tf.keras.layers.Flatten(),
        dense_block(512, 0.7),
        dense_block(128, 0.5),
        dense_block(64, 0.3),

        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    return model

#     ( )
def plot_train_history(history, title_prefix=""):
    fig, ax = plt.subplots(1, 4, figsize=(20, 3))
    ax = ax.ravel()

    for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
        ax[i].plot(history[met])
        ax[i].plot(history['val_' + met])
        ax[i].set_title(f'{title_prefix} Model {met}')
        ax[i].set_xlabel('epochs')
        ax[i].set_ylabel(met)
        ax[i].legend(['train', 'val'])
    plt.show()

#      (:   )
def save_model_results(model_obj, history_obj, model_name):
    model_dir = os.path.join(ROOT_PATH, 'saved_model')
    history_dir = os.path.join(ROOT_PATH, 'history')

    # Create directories if they don't exist
    os.makedirs(model_dir, exist_ok=True)
    os.makedirs(history_dir, exist_ok=True)

    model_path = os.path.join(model_dir, model_name)
    hist_path = os.path.join(history_dir, f'{model_name}_history.pkl')

    model_obj.save(model_path)
    joblib.dump(history_obj.history, hist_path)

def load_model_results(model_name):
    model_path = os.path.join(ROOT_PATH, f'saved_model/{model_name}')
    hist_path = os.path.join(ROOT_PATH, f'history/{model_name}_history.pkl')
    loaded_model = tf.keras.models.load_model(model_path)
    loaded_history = joblib.load(hist_path)
    return loaded_model, loaded_history

#    ( )
#    (res   )
theta_low = -0.3
theta_high = 0.3
hrz_shift = 40 * (res / 180) #   
ver_shift = 40 * (res / 180) #   
size_low = 0.8
size_high = 1.3
cont_low = 0.7
cont_high = 1.3

def apply_mat(image, transform_mat):
    rows = tf.range(-res/2, res/2)
    cols = tf.range(-res/2, res/2)

    coords = tf.stack([
        tf.repeat(rows, len(cols)),
        tf.tile(cols, [len(rows)])
    ], axis=1)

    trn_coords = tf.transpose(tf.linalg.matmul(transform_mat, tf.transpose(coords)))

    offset = tf.stack([
        tf.repeat(res/2, len(trn_coords)),
        tf.repeat(res/2, len(trn_coords))
    ], axis=1)

    trn_coords = tf.math.add(trn_coords, offset)
    trn_coords = tf.clip_by_value(trn_coords, 0, res-1)

    trn_pixels = tf.gather_nd(image, tf.cast(trn_coords, tf.int32))
    trn_image = tf.reshape(trn_pixels, [res, res, 3])
    return trn_image

def random_rotate(image):
    theta = tf.random.uniform([], minval=theta_low, maxval=theta_high)
    transform_mat = tf.convert_to_tensor([
        [tf.math.cos(theta), -tf.math.sin(theta)],
        [tf.math.sin(theta),  tf.math.cos(theta)]
    ])
    return apply_mat(image, transform_mat)

def random_shift(image):
    # Fix: Cast hrz_shift and ver_shift to int32 for tf.random.uniform maxval
    h1 = tf.random.uniform([], minval=0, maxval=tf.cast(hrz_shift, tf.int32), dtype=tf.int32)
    h2 = tf.random.uniform([], minval=0, maxval=tf.cast(hrz_shift, tf.int32), dtype=tf.int32)
    w1 = tf.random.uniform([], minval=0, maxval=tf.cast(ver_shift, tf.int32), dtype=tf.int32)
    w2 = tf.random.uniform([], minval=0, maxval=tf.cast(ver_shift, tf.int32), dtype=tf.int32)

    trn_image = tf.image.crop_to_bounding_box(image, h1, w1, res - h1 - h2, res - w1 - w2)
    trn_image = tf.image.resize(trn_image, size=IMAGE_SIZE, method="bicubic")
    return tf.reshape(trn_image, IMAGE_SIZE + [3])

def random_zoom(image):
    size_factor = tf.random.uniform([], minval=size_low, maxval=size_high)
    r_size = tf.cast(size_factor * res, tf.int32)
    s_image = tf.image.resize(image, size=[r_size, r_size], method="bicubic")
    trn_image = tf.image.resize_with_crop_or_pad(s_image, res, res)
    return tf.reshape(trn_image, [res, res, 3])

def random_contrast(image, max_val=1.0, min_val=0.0):
    cont_factor = tf.random.uniform([], minval=cont_low, maxval=cont_high)
    trn_image = tf.image.adjust_contrast(image, cont_factor)
    trn_image = tf.clip_by_value(trn_image, min_val, max_val)
    return tf.reshape(trn_image, [res, res, 3])

def augment(image, label):
    fn_idx = tf.random.uniform([], maxval=4, dtype=tf.int32)
    funcs = [random_rotate, random_shift, random_zoom, random_contrast]
    image = tf.switch_case(fn_idx, branch_fns=[lambda f=f: f(image) for f in funcs])
    return image, label

def prepare_for_training_with_augmentation(ds, shuffle_buffer_size=1000):
    ds = ds.map(augment, num_parallel_calls=AUTOTUNE) # augment 
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

def prepare_for_testing(ds, shuffle_buffer_size=1000): # test  augmentation  
    ds = ds.shuffle(buffer_size=shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

# --- Baseline     ---
#      .
# print("--- Baseline Model Training and Evaluation ---")
# IMAGE_SIZE = [180, 180]
# res = IMAGE_SIZE[0] # res          

# train_list_ds_baseline = tf.data.Dataset.from_tensor_slices(train_filenames)
# val_list_ds_baseline = tf.data.Dataset.from_tensor_slices(val_filenames)
# test_list_ds_baseline = tf.data.Dataset.list_files(TEST_PATH)

# train_ds_baseline = train_list_ds_baseline.map(process_path, num_parallel_calls=AUTOTUNE)
# val_ds_baseline = val_list_ds_baseline.map(process_path, num_parallel_calls=AUTOTUNE)
# test_ds_baseline = test_list_ds_baseline.map(process_path, num_parallel_calls=AUTOTUNE)

# TRAIN_IMG_COUNT_BASELINE = tf.data.experimental.cardinality(train_ds_baseline).numpy()
# VAL_IMG_COUNT_BASELINE = tf.data.experimental.cardinality(val_ds_baseline).numpy()
# TEST_IMAGE_COUNT_BASELINE = tf.data.experimental.cardinality(test_list_ds_baseline).numpy()

# train_ds_baseline = prepare_for_testing(train_ds_baseline) # baseline augment 
# val_ds_baseline = prepare_for_testing(val_ds_baseline) # baseline augment 
# test_ds_baseline = test_ds_baseline.batch(BATCH_SIZE) # test batch 

# weight_for_0_baseline = (1 / COUNT_NORMAL) * (TRAIN_IMG_COUNT_BASELINE) / 2.0
# weight_for_1_baseline = (1 / COUNT_PNEUMONIA) * (TRAIN_IMG_COUNT_BASELINE) / 2.0
# class_weight_baseline = {0: weight_for_0_baseline, 1: weight_for_1_baseline}

# with tf.device('/GPU:0'):
#     baseline_model = build_model()
#     METRICS = [
#         'accuracy',
#         tf.keras.metrics.Precision(name='precision'),
#         tf.keras.metrics.Recall(name='recall')
#     ]
#     baseline_model.compile(
#         optimizer='adam',
#         loss='binary_crossentropy',
#         metrics=METRICS
#     )
#     baseline_history = baseline_model.fit(
#         train_ds_baseline,
#         steps_per_epoch=TRAIN_IMG_COUNT_BASELINE // BATCH_SIZE,
#         epochs=EPOCHS,
#         validation_data=val_ds_baseline,
#         validation_steps=VAL_IMG_COUNT_BASELINE // BATCH_SIZE,
#         class_weight=class_weight_baseline,
#     )

# print("\n--- Baseline Model Evaluation Results ---")
# baseline_eval_results = baseline_model.evaluate(test_ds_baseline)
# print(f"Loss: {baseline_eval_results[0]:.4f}, Accuracy: {baseline_eval_results[1]:.4f}, "
#       f"Precision: {baseline_eval_results[2]:.4f}, Recall: {baseline_eval_results[3]:.4f}")

# plot_train_history(baseline_history.history, title_prefix="Baseline")
# save_model_results(baseline_model, baseline_history, 'baseline_model')


# --- Data Augmentation     ---
print("\n--- Data Augmentation Model Training and Evaluation ---")
# Change IMAGE_SIZE for data augmentation model
IMAGE_SIZE = [224, 224] # Change size for augmentation
res = IMAGE_SIZE[0] # res  IMAGE_SIZE  

#    IMAGE_SIZE   
theta_low = -0.3
theta_high = 0.3
hrz_shift = 40 * (res / 180) #   
ver_shift = 40 * (res / 180) #   
size_low = 0.8
size_high = 1.3
cont_low = 0.7
cont_high = 1.3


train_list_ds_aug = tf.data.Dataset.from_tensor_slices(train_filenames)
val_list_ds_aug = tf.data.Dataset.from_tensor_slices(val_filenames)
test_list_ds_aug = tf.data.Dataset.list_files(TEST_PATH)

train_ds_aug = train_list_ds_aug.map(process_path, num_parallel_calls=AUTOTUNE)
val_ds_aug = val_list_ds_aug.map(process_path, num_parallel_calls=AUTOTUNE)
test_ds_aug = test_list_ds_aug.map(process_path, num_parallel_calls=AUTOTUNE) # Test set no augmentation

TRAIN_IMG_COUNT_AUG = tf.data.experimental.cardinality(train_ds_aug).numpy()
VAL_IMG_COUNT_AUG = tf.data.experimental.cardinality(val_ds_aug).numpy()
TEST_IMAGE_COUNT_AUG = tf.data.experimental.cardinality(test_list_ds_aug).numpy()

train_ds_aug = prepare_for_training_with_augmentation(train_ds_aug) # Apply augmentation
val_ds_aug = prepare_for_testing(val_ds_aug) # Validation set no augmentation
test_ds_aug = test_ds_aug.batch(BATCH_SIZE)

weight_for_0_aug = (1 / COUNT_NORMAL) * (TRAIN_IMG_COUNT_AUG) / 2.0
weight_for_1_aug = (1 / COUNT_PNEUMONIA) * (TRAIN_IMG_COUNT_AUG) / 2.0
class_weight_aug = {0: weight_for_0_aug, 1: weight_for_1_aug}

with tf.device('/GPU:0'):
    aug_model = build_model()
    METRICS = [
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
    aug_model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=METRICS
    )
    aug_history = aug_model.fit(
        train_ds_aug,
        steps_per_epoch=TRAIN_IMG_COUNT_AUG // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=val_ds_aug,
        validation_steps=VAL_IMG_COUNT_AUG // BATCH_SIZE,
        class_weight=class_weight_aug,
    )

print("\n--- Data Augmentation Model Evaluation Results ---")
aug_eval_results = aug_model.evaluate(test_ds_aug)
print(f"Loss: {aug_eval_results[0]:.4f}, Accuracy: {aug_eval_results[1]:.4f}, "
      f"Precision: {aug_eval_results[2]:.4f}, Recall: {aug_eval_results[3]:.4f}")

plot_train_history(aug_history.history, title_prefix="Data Augmentation")
save_model_results(aug_model, aug_history, 'aug_model')


# ---   ---
print("\n--- Final Comparison ---")
#       
# Baseline   , Data Augmentation    ,
# Baseline        .
#  Baseline  history ,     .
#  Baseline  history    load_model_results  .

# : Baseline  history    (  )
try:
    # baseline_model, baseline_history_loaded = load_model_results('baseline_model')
    # plot_train_history     history  .
    baseline_history_loaded = joblib.load(os.path.join(ROOT_PATH, 'history/baseline_model_history.pkl'))

    plt.figure(figsize=(20, 4))
    for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
        plt.subplot(1, 4, i + 1)
        plt.plot(baseline_history_loaded[met], label='Baseline Train')
        plt.plot(baseline_history_loaded['val_' + met], label='Baseline Val')
        plt.plot(aug_history.history[met], label='Augmented Train')
        plt.plot(aug_history.history['val_' + met], label='Augmented Val')
        plt.title(f'Comparison of {met}')
        plt.xlabel('Epochs')
        plt.ylabel(met)
        plt.legend()
    plt.tight_layout()
    plt.show()

    print("\n--- Summary of Model Performance ---")
    # Baseline     .       ,
    #     .
    # baseline_eval_results_loaded = baseline_model.evaluate(test_ds_baseline) #   test_ds_baseline 
    print(f"Baseline Model (Loaded): Accuracy={baseline_eval_results[1]:.4f}, Precision={baseline_eval_results[2]:.4f}, Recall={baseline_eval_results[3]:.4f}")
    print(f"Augmented Model: Accuracy={aug_eval_results[1]:.4f}, Precision={aug_eval_results[2]:.4f}, Recall={aug_eval_results[3]:.4f}")

except FileNotFoundError:
    print("\nBaseline model history not found. Skipping final comparison plot with Baseline model.")
    print("\n--- Summary of Data Augmentation Model Performance ---")
    print(f"Augmented Model: Accuracy={aug_eval_results[1]:.4f}, Precision={aug_eval_results[2]:.4f}, Recall={aug_eval_results[3]:.4f}")
```
