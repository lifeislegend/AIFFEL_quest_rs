## Overview

Generative image models learn a "latent manifold" of the visual world:
a low-dimensional vector space where each point maps to an image.
Going from such a point on the manifold back to a displayable image
is called "decoding" -- in the Stable Diffusion model, this is handled by
the "decoder" model.

![The Stable Diffusion architecture](https://i.imgur.com/2uC8rYJ.png)

This latent manifold of images is continuous and interpolative, meaning that:

1. Moving a little on the manifold only changes the corresponding image a little (continuity).
2. For any two points A and B on the manifold (i.e. any two images), it is possible
to move from A to B via a path where each intermediate point is also on the manifold (i.e.
is also a valid image). Intermediate points would be called "interpolations" between
the two starting images.

Stable Diffusion isn't just an image model, though, it's also a natural language model.
It has two latent spaces: the image representation space learned by the
encoder used during training, and the prompt latent space
which is learned using a combination of pretraining and training-time
fine-tuning.

_Latent space walking_, or _latent space exploration_, is the process of
sampling a point in latent space and incrementally changing the latent
representation. Its most common application is generating animations
where each sampled point is fed to the decoder and is stored as a
frame in the final animation.
For high-quality latent representations, this produces coherent-looking
animations. These animations can provide insight into the feature map of the
latent space, and can ultimately lead to improvements in the training
process. One such GIF is displayed below:

![Panda to Plane](https://keras.io/img/examples/generative/random_walks_with_stable_diffusion/panda2plane.gif)

In this guide, we will show how to take advantage of the Stable Diffusion API
in KerasCV to perform prompt interpolation and circular walks through
Stable Diffusion's visual latent manifold, as well as through
the text encoder's latent manifold.

## Interpolating between text prompts

In Stable Diffusion, a text prompt is first encoded into a vector,
and that encoding is used to guide the diffusion process.
The latent encoding vector has shape
77x768 (that's huge!), and when we give Stable Diffusion a text prompt, we're
generating images from just one such point on the latent manifold.

To explore more of this manifold, we can interpolate between two text encodings
and generate images at those interpolated points:

````python
# 0. ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
import tensorflow as tf
import keras_cv
import keras
# import matplotlib.pyplot as plt # Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî Ïãú ÌïÑÏöî
# import math # Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî Ïãú ÌïÑÏöî

print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {keras.__version__}")
print(f"KerasCV version: {keras_cv.__version__}")

# 1. Keras Mixed Precision Ï†ïÏ±Ö ÏÑ§Ï†ï (Í∞ÄÏû• ÏïàÏ†ïÏ†ÅÏù∏ float32Î°ú ÏÑ§Ï†ï)
keras.mixed_precision.set_global_policy("float32")
print("Using Keras mixed precision policy: float32")

# 2. Stable Diffusion Î™®Îç∏ Ïù∏Ïä§ÌÑ¥Ïä§Ìôî (jit_compile=FalseÎ°ú ÏÑ§Ï†ï)
# Ïù¥ Î∂ÄÎ∂ÑÏóêÏÑú Î™®Îç∏ Í∞ÄÏ§ëÏπò Îã§Ïö¥Î°úÎìúÍ∞Ä Î∞úÏÉùÌï† Ïàò ÏûàÏäµÎãàÎã§.
try:
    print("Instantiating StableDiffusion model with jit_compile=False...")
    model = keras_cv.models.StableDiffusion(jit_compile=False)
    print("StableDiffusion model instantiated successfully.")
except Exception as e:
    print(f"Error during model instantiation: {e}")
    # Î™®Îç∏ Ïù∏Ïä§ÌÑ¥Ïä§Ìôî Ïã§Ìå® Ïãú Ïó¨Í∏∞ÏÑú Ï§ëÎã®
    raise

# 3. ÌÖçÏä§Ìä∏ ÌîÑÎ°¨ÌîÑÌä∏ Ï†ïÏùò
prompt_1 = "A watercolor painting of a Golden Retriever at the beach"
prompt_2 = "A still life DSLR photo of a bowl of fruit"
prompt_3 = "The eiffel tower in the style of starry night"
prompt_4 = "An architectural sketch of a skyscraper"

interpolation_steps = 5

# 4. ÌÖçÏä§Ìä∏ ÌîÑÎ°¨ÌîÑÌä∏ Ïù∏ÏΩîÎî© ÏãúÎèÑ
try:
    print("Encoding text prompts...")
    encoding_1 = tf.squeeze(model.encode_text(prompt_1))
    encoding_2 = tf.squeeze(model.encode_text(prompt_2))
    encoding_3 = tf.squeeze(model.encode_text(prompt_3))
    encoding_4 = tf.squeeze(model.encode_text(prompt_4))
    print("Text prompts encoded successfully.")

    # Ïù∏ÏΩîÎî©Îêú ÌÖêÏÑúÏùò ÌòïÌÉú Ï∂úÎ†•
    print(f"Encoding shape (e.g., encoding_1): {encoding_1.shape}")

    # tf.linspace ÏÇ¨Ïö©Î≤ï ÏàòÏ†ï (Ïòà: encoding_1Í≥º encoding_2 ÏÇ¨Ïù¥ Î≥¥Í∞Ñ)
    interpolated_encodings_1_to_2 = tf.linspace(encoding_1, encoding_2, interpolation_steps)
    print(f"Shape of interpolated encodings (1 to 2): {interpolated_encodings_1_to_2.shape}")

except Exception as e:
    print(f"Error during text encoding or linspace: {e}")
    # Ïò§Î•ò Î∞úÏÉù Ïãú ÏÉÅÏÑ∏ Ï†ïÎ≥¥ Ï∂úÎ†•
    import traceback
    traceback.print_exc()

!pip uninstall -y keras-cv tensorflow keras
!pip install tensorflow==2.15.0 keras==2.15.0 keras-cv==0.8.2

````

Once we've interpolated the encodings, we can generate images from each point.
Note that in order to maintain some stability between the resulting images we
keep the diffusion noise constant between images.

````python
import tensorflow as tf
import keras_cv
import keras
import matplotlib.pyplot as plt
import math

# --- Cell 1 ÏóêÏÑú model Í≥º interpolated_encodings_1_to_2 Í∞Ä Ï†ïÏùòÎêòÏóàÎã§Í≥† Í∞ÄÏ†ï ---
# model = keras_cv.models.StableDiffusion(jit_compile=False)
# encoding_1 = tf.squeeze(model.encode_text(prompt_1)) # Cell 1Ïùò Í≤∞Í≥º Ï§ë Ï≤´ Î≤àÏß∏ Ïù∏ÏΩîÎî©

# Cell 2 ÏãúÏûë (Îã®Ïùº Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÌÖåÏä§Ìä∏)

# 0. ÏÇ¨Ïö©Ìï† Îã®Ïùº Ïù∏ÏΩîÎî© (Ïòà: Î≥¥Í∞ÑÎêú Í≤∞Í≥º Ï§ë Ï≤´ Î≤àÏß∏ ÎòêÎäî ÏõêÎ≥∏ Ïù∏ÏΩîÎî©_1)
#    interpolated_encodings_1_to_2[0]Ïùò shapeÏùÄ (77, 768)Ïù¥ Îê©ÎãàÎã§.
#    generate_imageÎäî (batch_size, 77, 768)ÏùÑ Í∏∞ÎåÄÌïòÎØÄÎ°ú expand_dims ÌïÑÏöî
single_encoding_for_generation = tf.expand_dims(interpolated_encodings_1_to_2[0], axis=0) # Shape: (1, 77, 768)

seed = 12345

# 1. Îã®Ïùº ÎÖ∏Ïù¥Ï¶à ÏÉùÏÑ±
#    diffusion_noiseÎäî generate_imageÏóê Ï†ÑÎã¨Îê† Îïå (batch_size, height/8, width/8, 4) ÌòïÌÉúÎ•º Í∏∞ÎåÄÌï©ÎãàÎã§.
#    Ïó¨Í∏∞ÏÑúÎäî batch_size=1 ÏûÖÎãàÎã§.
single_diffusion_noise = tf.random.normal(
    (1, 512 // 8, 512 // 8, 4), # (1, 64, 64, 4)
    seed=seed
)

# 2. Negative prompt ÏÑ§Ï†ï
negative_prompt_for_single = "" # Îã®Ïùº Î¨∏ÏûêÏó¥

print(f"Generating 1 image...")
print(f"Shape of single_encoding_for_generation: {single_encoding_for_generation.shape}")
print(f"Shape of single_diffusion_noise: {single_diffusion_noise.shape}")

try:
    image_batch = model.generate_image(
        encoded_text=single_encoding_for_generation, # Î∞∞Ïπò ÌÅ¨Í∏∞ 1Ïù∏ Ïù∏ÏΩîÎî©
        negative_prompt=negative_prompt_for_single,
        num_steps=25,
        diffusion_noise=single_diffusion_noise, # Î∞∞Ïπò ÌÅ¨Í∏∞ 1Ïù∏ ÎÖ∏Ïù¥Ï¶à
    )
    print("Image generated successfully!")
    image = image_batch[0] # Î∞∞Ïπò Ï§ë Ï≤´ Î≤àÏß∏ (Ïù¥Ïûê Ïú†ÏùºÌïú) Ïù¥ÎØ∏ÏßÄ

    # ÏÉùÏÑ±Îêú Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title("Single Generated Image")
    plt.axis("off")
    plt.show()

except Exception as e:
    print(f"Error during single image generation: {e}")
    import traceback
    traceback.print_exc()
````

````python
# Ïù¥Ï†Ñ ÏÖÄÏóêÏÑú 'image' Î≥ÄÏàòÏóê Îã®Ïùº Ïù¥ÎØ∏ÏßÄÍ∞Ä numpy Î∞∞Ïó¥Î°ú ÏÑ±Í≥µÏ†ÅÏúºÎ°ú ÏÉùÏÑ±ÎêòÏóàÎã§Í≥† Í∞ÄÏ†ïÌï©ÎãàÎã§.
# Ïòà: image = image_batch[0] ÏôÄ Í∞ôÏù¥Ïöî.
# ÎßåÏïΩ 'image' Î≥ÄÏàòÍ∞Ä Ï†ïÏùòÎêòÏßÄ ÏïäÏïòÎã§Î©¥, Ïù¥ ÏΩîÎìú Ïã§Ìñâ Ï†ÑÏóê Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÏÖÄÏùÑ Îã§Ïãú Ïã§ÌñâÌï¥Ïïº Ìï©ÎãàÎã§.

from PIL import Image # PIL.Image Î™®ÎìàÏùÑ ÏûÑÌè¨Ìä∏Ìï©ÎãàÎã§.
# from IPython.display import Image as IImage # GIF ÌëúÏãúÎ•º ÏúÑÌï¥ ÌïÑÏöîÌï† Ïàò ÏûàÏäµÎãàÎã§.

# export_as_gif Ìï®Ïàò Ï†ïÏùò (Ïù¥ÎØ∏ Ï†ïÏùòÌñàÎã§Î©¥ Îã§Ïãú Ï†ïÏùòÌï† ÌïÑÏöîÎäî ÏóÜÏùÑ Ïàò ÏûàÏäµÎãàÎã§)
def export_as_gif(filename, pil_image_list, frames_per_second=10, rubber_band=False):
    """
    PIL Image Í∞ùÏ≤¥Îì§Ïùò Î¶¨Ïä§Ìä∏Î•º GIF ÌååÏùºÎ°ú Ï†ÄÏû•Ìï©ÎãàÎã§.
    pil_image_list: PIL.Image Í∞ùÏ≤¥Îì§Ïùò Î¶¨Ïä§Ìä∏Ïó¨Ïïº Ìï©ÎãàÎã§.
    """
    if not pil_image_list:
        print("Ï†ÄÏû•Ìï† Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏóÜÏäµÎãàÎã§.")
        return

    # rubber_band Î°úÏßÅ ÏàòÏ†ï (Ïù¥ÎØ∏ÏßÄÍ∞Ä Ï∂©Î∂ÑÌûà ÏûàÏùÑ ÎïåÎßå Ï†ÅÏö©)
    if rubber_band and len(pil_image_list) >= 2: # ÏµúÏÜå 2Ïû• Ïù¥ÏÉÅÏùº Îïå ÏùòÎØ∏Í∞Ä ÏûàÏùå
        # Ïòà: A, B, C -> A, B, C, B (ÎßàÏßÄÎßâ Ï†úÏô∏ÌïòÍ≥† Îí§ÏßëÏñ¥ÏÑú Ï∂îÍ∞Ä)
        # ÎòêÎäî A, B, C, D -> A, B, C, D, C, B
        to_append_reversed = pil_image_list[1:-1] # Ï≤òÏùåÍ≥º ÎßàÏßÄÎßâ Ï†úÏô∏ (ÎòêÎäî ÎßàÏßÄÎßâÎßå Ï†úÏô∏ Îì±)
        if to_append_reversed: # Ï∂îÍ∞ÄÌï† ÌîÑÎ†àÏûÑÏù¥ ÏûàÎã§Î©¥
             pil_image_list += to_append_reversed[::-1]
        # ÏõêÎûò ÏΩîÎìú: images += images[2:-1][::-1]
        # Ïù¥ Î∂ÄÎ∂ÑÏùÄ Ïù¥ÎØ∏ÏßÄ Í∞úÏàòÏóê Îî∞Îùº Îã§Î•¥Í≤å ÎèôÏûëÌï† Ïàò ÏûàÏúºÎØÄÎ°ú, ÏùòÎèÑÏóê ÎßûÍ≤å ÏàòÏ†ï ÌïÑÏöî Ïãú ÏïåÎ†§Ï£ºÏÑ∏Ïöî.
        # ÌòÑÏû¨Îäî Í∞ÑÎã®Ìûà Îëê Î≤àÏß∏Î∂ÄÌÑ∞ ÎßàÏßÄÎßâÏóêÏÑú Îëê Î≤àÏß∏ ÌîÑÎ†àÏûÑÍπåÏßÄÎ•º Îí§ÏßëÏñ¥ Ï∂îÍ∞ÄÌïòÎäî Î∞©ÏãùÏúºÎ°ú Í∞ÄÏ†ïÌï©ÎãàÎã§.
        # Îçî Í∞ÑÎã®ÌïòÍ≤åÎäî: if len(pil_image_list) > 1: pil_image_list += pil_image_list[:-1][::-1] (ÎßàÏßÄÎßâ Ï†úÏô∏ÌïòÍ≥† Ï†ÑÎ∂Ä Îí§ÏßëÏñ¥ Ï∂îÍ∞Ä)

    pil_image_list[0].save(
        filename,
        save_all=True,
        append_images=pil_image_list[1:],
        duration=1000 // frames_per_second,
        loop=0,
    )
    print(f"GIF ÌååÏùºÏù¥ '{filename}'ÏúºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")


# --- Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± Î∞è GIF Î≥ÄÌôò ---
try:
    # 1. Ïù¥Ï†ÑÏóê ÏÉùÏÑ±Îêú Îã®Ïùº Ïù¥ÎØ∏ÏßÄ(numpy Î∞∞Ïó¥)Î•º 'image' Î≥ÄÏàòÏóêÏÑú Í∞ÄÏ†∏ÏòµÎãàÎã§.
    #    ÎßåÏïΩ 'image' Î≥ÄÏàòÍ∞Ä ÏóÜÎã§Î©¥ NameErrorÍ∞Ä Ïó¨Í∏∞ÏÑú Î∞úÏÉùÌï©ÎãàÎã§.
    #    Ïù¥ Í≤ΩÏö∞, Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÏΩîÎìúÎ•º Î®ºÏ†Ä Ïã§ÌñâÌï¥Ïïº Ìï©ÎãàÎã§.
    #    ÏòàÎ•º Îì§Ïñ¥, Ïù¥Ï†Ñ Îã®Í≥ÑÏóêÏÑú image = image_batch[0] ÏôÄ Í∞ôÏù¥ Îã®Ïùº Ïù¥ÎØ∏ÏßÄÎ•º Ï†ÄÏû•ÌñàÏäµÎãàÎã§.

    # 'image' Î≥ÄÏàòÍ∞Ä Ï†ïÏùòÎêòÏñ¥ ÏûàÎã§Í≥† Í∞ÄÏ†ïÌïòÍ≥†, Ïù¥Î•º Î¶¨Ïä§Ìä∏Ïóê ÎÑ£ÏäµÎãàÎã§.
    # 'image'Îäî numpy Î∞∞Ïó¥Ïù¥Ïñ¥Ïïº Ìï©ÎãàÎã§.
    numpy_image_list_for_gif = [image]

    # 2. Numpy Î∞∞Ïó¥ Î¶¨Ïä§Ìä∏Î•º PIL Image Í∞ùÏ≤¥ Î¶¨Ïä§Ìä∏Î°ú Î≥ÄÌôòÌï©ÎãàÎã§.
    pil_images = [Image.fromarray(img_arr) for img_arr in numpy_image_list_for_gif]

    # 3. GIFÎ°ú Ï†ÄÏû•Ìï©ÎãàÎã§.
    export_as_gif(
        "doggo-and-fruit-single.gif", # ÌååÏùº Ïù¥Î¶ÑÏùÑ Îã§Î•¥Í≤å Ìï¥ÏÑú ÌÖåÏä§Ìä∏Ìï¥Î≥¥ÏÑ∏Ïöî.
        pil_images,
        frames_per_second=2, # Ï¥àÎãπ ÌîÑÎ†àÏûÑ Ïàò
        rubber_band=True, # rubber_band Ìö®Í≥ºÎ•º Ï£ºÎ†§Î©¥ Ïù¥ÎØ∏ÏßÄÍ∞Ä Ïó¨Îü¨ Ïû• ÌïÑÏöîÌï©ÎãàÎã§.
                           # ÌòÑÏû¨Îäî Ïù¥ÎØ∏ÏßÄÍ∞Ä Ìïú Ïû•Ïù¥ÎØÄÎ°ú rubber_band Ìö®Í≥ºÎäî ÌÅ¨Í≤å ÎÇòÌÉÄÎÇòÏßÄ ÏïäÏäµÎãàÎã§.
    )

    # 4. (ÏÑ†ÌÉù ÏÇ¨Ìï≠) ÏÉùÏÑ±Îêú GIFÎ•º ÎÖ∏Ìä∏Î∂ÅÏóê ÌëúÏãúÌï©ÎãàÎã§.
    # from IPython.display import Image as IImage
    # display(IImage(filename="doggo-and-fruit-single.gif"))

except NameError:
    print("Ïò§Î•ò: 'image' Î≥ÄÏàòÍ∞Ä Ï†ïÏùòÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§. Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÏÖÄÏùÑ Î®ºÏ†Ä Ïã§ÌñâÌïòÍ≥†,")
    print("ÏÉùÏÑ±Îêú Îã®Ïùº Ïù¥ÎØ∏ÏßÄÍ∞Ä 'image'ÎùºÎäî Î≥ÄÏàòÏóê Ï†ÄÏû•ÎêòÏñ¥ ÏûàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
except Exception as e:
    print(f"GIF ÏÉùÏÑ± Ï§ë Ïò§Î•ò Î∞úÏÉù: {e}")
    import traceback
    traceback.print_exc()
````

The results may seem surprising. Generally, interpolating between prompts
produces coherent looking images, and often demonstrates a progressive concept
shift between the contents of the two prompts. This is indicative of a high
quality representation space, that closely mirrors the natural structure
of the visual world.

To best visualize this, we should do a much more fine-grained interpolation,
using hundreds of steps. In order to keep batch size small (so that we don't
OOM our GPU), this requires manually batching our interpolated
encodings.

````python
import tensorflow as tf
import keras_cv # Ïù¥ÎØ∏ ÏûÑÌè¨Ìä∏ ÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï®
import keras    # Ïù¥ÎØ∏ ÏûÑÌè¨Ìä∏ ÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï®
from PIL import Image # GIF ÏÉùÏÑ±ÏùÑ ÏúÑÌï¥ ÌïÑÏöî
# import matplotlib.pyplot as plt # Í∞Å Ïù¥ÎØ∏ÏßÄ ÌôïÏù∏ Ïãú ÌïÑÏöîÌï† Ïàò ÏûàÏùå
# import math # ÌïÑÏöîÏãú ÏÇ¨Ïö©

# --- Ïù¥Ï†Ñ ÏÖÄÏóêÏÑú model, encoding_1, encoding_2 Î≥ÄÏàòÍ∞Ä ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ï†ïÏùòÎêòÏóàÎã§Í≥† Í∞ÄÏ†ï ---
# Ïòà (Ïù¥Îü∞ ÏΩîÎìúÎì§Ïù¥ Ïù¥Ï†Ñ ÏÖÄÏóê ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§):
# print(f"TensorFlow version: {tf.__version__}")
# print(f"Keras version: {keras.__version__}")
# print(f"KerasCV version: {keras_cv.__version__}")
# keras.mixed_precision.set_global_policy("float32")
# model = keras_cv.models.StableDiffusion(jit_compile=False) # model Ï†ïÏùò
# prompt_1 = "A watercolor painting of a Golden Retriever at the beach"
# prompt_2 = "A still life DSLR photo of a bowl of fruit"
# encoding_1 = tf.squeeze(model.encode_text(prompt_1)) # encoding_1 Ï†ïÏùò
# encoding_2 = tf.squeeze(model.encode_text(prompt_2)) # encoding_2 Ï†ïÏùò
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---

# export_as_gif Ìï®Ïàò Ï†ïÏùò (Ïù¥ÎØ∏ ÌòÑÏû¨ ÏÑ∏ÏÖòÏóê Ï†ïÏùòÎêòÏñ¥ ÏûàÎã§Î©¥ Îã§Ïãú Ï†ïÏùòÌï† ÌïÑÏöî ÏóÜÏùå)
def export_as_gif(filename, pil_image_list, frames_per_second=10, rubber_band=False):
    if not pil_image_list:
        print("Ï†ÄÏû•Ìï† Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏóÜÏäµÎãàÎã§.")
        return
    if rubber_band and len(pil_image_list) >= 2:
        to_append_reversed = pil_image_list[1:-1] # Ï≤òÏùåÍ≥º ÎßàÏßÄÎßâ ÌîÑÎ†àÏûÑ Ï†úÏô∏
        if to_append_reversed:
             pil_image_list += to_append_reversed[::-1] # Îí§ÏßëÏñ¥ÏÑú Ï∂îÍ∞Ä

    pil_image_list[0].save(
        filename,
        save_all=True,
        append_images=pil_image_list[1:],
        duration=1000 // frames_per_second,
        loop=0,
    )
    print(f"GIF ÌååÏùºÏù¥ '{filename}'ÏúºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")

# --- 150Í∞ú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± Î∞è GIF Î≥ÄÌôò ÏãúÏûë (Îã®Ïùº Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± Î∞òÎ≥µ) ---
try:
    interpolation_steps = 6

    # 1. ÌÖçÏä§Ìä∏ Ïù∏ÏΩîÎî© Î≥¥Í∞Ñ (tf.linspace ÏÇ¨Ïö©)
    print(f"Interpolating text encodings for {interpolation_steps} steps...")
    # encoding_1Í≥º encoding_2Í∞Ä ÌòÑÏû¨ Î≤îÏúÑÏóê Ï†ïÏùòÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§.
    interpolated_encodings = tf.linspace(encoding_1, encoding_2, interpolation_steps) # Shape: (10, 77, 768)
    print(f"Shape of interpolated_encodings: {interpolated_encodings.shape}")

    all_generated_pil_images = []
    base_seed = 12345 # ÎÖ∏Ïù¥Ï¶à ÏÉùÏÑ±ÏùÑ ÏúÑÌïú Í∏∞Î≥∏ ÏãúÎìú
    negative_prompt_for_iterations = "" # Î™®Îì† Î∞òÎ≥µÏóê ÎèôÏùºÌïú negative_prompt ÏÇ¨Ïö©

    print(f"\nStarting iterative image generation for {interpolation_steps} images...")
    for i in range(interpolation_steps):
        print(f"  Generating image {i+1}/{interpolation_steps}...")

        current_single_encoding = tf.expand_dims(interpolated_encodings[i], axis=0)
        current_single_noise = tf.random.normal(
            (1, 512 // 8, 512 // 8, 4),
            seed=base_seed + i
        )

        # print(f"    Shape of current_single_encoding: {current_single_encoding.shape}")
        # print(f"    Shape of current_single_noise: {current_single_noise.shape}")

        # <<< Îì§Ïó¨Ïì∞Í∏∞ ÏàòÏ†ï: try Î∏îÎ°ùÏùÑ for Î£®ÌîÑ ÎÇ¥Ïùò Îã§Î•∏ ÏΩîÎìúÏôÄ ÎèôÏùºÌïú ÏàòÏ§ÄÏúºÎ°ú ÎßûÏ∂§ >>>
        try:
            generated_image_batch_np = model.generate_image(
                encoded_text=current_single_encoding,
                num_steps=25,
                diffusion_noise=current_single_noise,
                negative_prompt=negative_prompt_for_iterations,
            )
            single_image_np = generated_image_batch_np[0] # .numpy() Ï†úÍ±∞Îê®

            all_generated_pil_images.append(Image.fromarray(single_image_np))
            print(f"    Image {i+1} generated successfully.")

        except Exception as e_inner:
            print(f"    Error generating image {i+1}: {e_inner}")
            import traceback
            traceback.print_exc()
            pass

    print("\nAll image generation iterations complete.")

    if all_generated_pil_images:
        export_as_gif("doggo-and-fruit-150.gif", all_generated_pil_images, frames_per_second=10, rubber_band=True)
    else:
        print("No images were successfully generated to create a GIF.")

except NameError as ne:
    print(f"NameError Î∞úÏÉù: {ne}")
    print("Ïò§Î•òÍ∞Ä Î∞úÏÉùÌïú Î≥ÄÏàò(Ïòà: 'model', 'encoding_1', 'encoding_2')Í∞Ä ÌòÑÏû¨ ÏÖÄ Ïã§Ìñâ Ï†ÑÏóê Ï†ïÏùòÎêòÏóàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    print("ÌïÑÏöîÌïòÎã§Î©¥ Ìï¥Îãπ Î≥ÄÏàòÎì§ÏùÑ Ï†ïÏùòÌïòÎäî Ïù¥Ï†Ñ ÏÖÄÏùÑ Îã§Ïãú Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.")
    import traceback
    traceback.print_exc()
except Exception as e:
    print(f"An error occurred: {e}")
    import traceback
    traceback.print_exc()
````

## A walk around a text prompt

Our next experiment will be to go for a walk around the latent manifold
starting from a point produced by a particular prompt.

Ïù¥ Î∏îÎü≠ÏùÄ 15Î∂Ñ Ï†ïÎèÑ ÏÜåÏöîÎê©ÎãàÎã§!
Î∞∞ÏπòÏùò ÌÅ¨Í∏∞Î•º Ï§ÑÏó¨ÏÑú ÌÖåÏä§Ìä∏Ìï¥Î≥¥Í∏∏ Ï∂îÏ≤úÎìúÎ¶¥Í≤åÏöî!
ÎÖ∏Ìä∏Î∂ÅÏùò ÌååÏùºÏùÄ Í∏∞Ï°¥Ïùò Í≤∞Í≥ºÎ¨ºÎ°ú, ÏÉàÎ°ú ÌÖåÏä§Ìä∏ÌïòÎäî Í≤∞Í≥ºÎ¨ºÍ≥º Î∞∞Ïπò ÌÅ¨Í∏∞Ïóê Îî∞Î•∏ Í≤∞Í≥ºÎ•º ÎπÑÍµêÌï¥Î≥¥Îäî Í≤ÉÎèÑ Ï¢ãÏùÄ Ïã§ÌóòÏù¥Í≤†ÎÑ§Ïöî üòÄ

````python

walk_steps = 150
batch_size = 2
batches = walk_steps // batch_size
step_size = 0.005

encoding = tf.squeeze(
    model.encode_text("The Eiffel Tower in the style of starry night")
)
# Note that (77, 768) is the shape of the text encoding.
delta = tf.ones_like(encoding) * step_size

walked_encodings = []
for step_index in range(walk_steps):
    walked_encodings.append(encoding)
    encoding += delta
walked_encodings = tf.stack(walked_encodings)
batched_encodings = tf.split(walked_encodings, batches)

images = []
for batch in range(batches):
    images += [
        Image.fromarray(img)
        for img in model.generate_image(
            batched_encodings[batch],
            batch_size=batch_size,
            num_steps=25,
            diffusion_noise=noise,
        )
    ]

export_as_gif("eiffel-tower-starry-night.gif", images, rubber_band=True)

#15Î∂Ñ Ï†ïÎèÑ ÏÜåÏöîÎê©ÎãàÎã§!
````

## A circular walk through the diffusion noise space for a single prompt

Our final experiment is to stick to one prompt and explore the variety of images
that the diffusion model can produce from that prompt. We do this by controlling
the noise that is used to seed the diffusion process.

We create two noise components, `x` and `y`, and do a walk from 0 to 2œÄ, summing
the cosine of our `x` component and the sin of our `y` component to produce noise.
Using this approach, the end of our walk arrives at the same noise inputs where
we began our walk, so we get a "loopable" result!

````python
prompt = "An oil paintings of cows in a field next to a windmill in Holland"
encoding = tf.squeeze(model.encode_text(prompt))
walk_steps = 150
batch_size = 3
batches = walk_steps // batch_size

walk_noise_x = tf.random.normal(noise.shape, dtype=tf.float64)
walk_noise_y = tf.random.normal(noise.shape, dtype=tf.float64)

walk_scale_x = tf.cos(tf.linspace(0, 2, walk_steps) * math.pi)
walk_scale_y = tf.sin(tf.linspace(0, 2, walk_steps) * math.pi)
noise_x = tf.tensordot(walk_scale_x, walk_noise_x, axes=0)
noise_y = tf.tensordot(walk_scale_y, walk_noise_y, axes=0)
noise = tf.add(noise_x, noise_y)
batched_noise = tf.split(noise, batches)
````

Ïù¥ Î∏îÎü≠ÏùÄ 25Î∂Ñ Ï†ïÎèÑ ÏÜåÏöîÎê©ÎãàÎã§!
Î∞∞ÏπòÏùò ÌÅ¨Í∏∞Î•º Ï§ÑÏó¨ÏÑú ÌÖåÏä§Ìä∏Ìï¥Î≥¥Í∏∏ Ï∂îÏ≤úÎìúÎ¶¥Í≤åÏöî!

````python
#Ï∂îÍ∞Ä
encoding = tf.cast(encoding,tf.dtypes.float16)


images = []
for batch in range(batches):
    images += [
        Image.fromarray(img)
        for img in model.generate_image(
            #ÌòïÎ≥ÄÌôò
            tf.cast(encoding,tf.dtypes.float32),
            batch_size=batch_size,
            num_steps=25,
            #ÌòïÎ≥ÄÌôò
            diffusion_noise=tf.cast(batched_noise[batch],tf.dtypes.float16),
        )
    ]

export_as_gif("cows.gif", images)

#25Î∂Ñ Ï†ïÎèÑ ÏÜåÏöîÎê©ÎãàÎã§!
````
