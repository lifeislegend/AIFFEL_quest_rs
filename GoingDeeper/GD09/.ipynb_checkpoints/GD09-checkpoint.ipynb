## Overview

Generative image models learn a "latent manifold" of the visual world:
a low-dimensional vector space where each point maps to an image.
Going from such a point on the manifold back to a displayable image
is called "decoding" -- in the Stable Diffusion model, this is handled by
the "decoder" model.

![The Stable Diffusion architecture](https://i.imgur.com/2uC8rYJ.png)

This latent manifold of images is continuous and interpolative, meaning that:

1. Moving a little on the manifold only changes the corresponding image a little (continuity).
2. For any two points A and B on the manifold (i.e. any two images), it is possible
to move from A to B via a path where each intermediate point is also on the manifold (i.e.
is also a valid image). Intermediate points would be called "interpolations" between
the two starting images.

Stable Diffusion isn't just an image model, though, it's also a natural language model.
It has two latent spaces: the image representation space learned by the
encoder used during training, and the prompt latent space
which is learned using a combination of pretraining and training-time
fine-tuning.

_Latent space walking_, or _latent space exploration_, is the process of
sampling a point in latent space and incrementally changing the latent
representation. Its most common application is generating animations
where each sampled point is fed to the decoder and is stored as a
frame in the final animation.
For high-quality latent representations, this produces coherent-looking
animations. These animations can provide insight into the feature map of the
latent space, and can ultimately lead to improvements in the training
process. One such GIF is displayed below:

![Panda to Plane](https://keras.io/img/examples/generative/random_walks_with_stable_diffusion/panda2plane.gif)

In this guide, we will show how to take advantage of the Stable Diffusion API
in KerasCV to perform prompt interpolation and circular walks through
Stable Diffusion's visual latent manifold, as well as through
the text encoder's latent manifold.

## Interpolating between text prompts

In Stable Diffusion, a text prompt is first encoded into a vector,
and that encoding is used to guide the diffusion process.
The latent encoding vector has shape
77x768 (that's huge!), and when we give Stable Diffusion a text prompt, we're
generating images from just one such point on the latent manifold.

To explore more of this manifold, we can interpolate between two text encodings
and generate images at those interpolated points:

````python
# 0. 필요한 라이브러리 임포트
import tensorflow as tf
import keras_cv
import keras
# import matplotlib.pyplot as plt # 이미지 시각화 시 필요
# import math # 이미지 시각화 시 필요

print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {keras.__version__}")
print(f"KerasCV version: {keras_cv.__version__}")

# 1. Keras Mixed Precision 정책 설정 (가장 안정적인 float32로 설정)
keras.mixed_precision.set_global_policy("float32")
print("Using Keras mixed precision policy: float32")

# 2. Stable Diffusion 모델 인스턴스화 (jit_compile=False로 설정)
# 이 부분에서 모델 가중치 다운로드가 발생할 수 있습니다.
try:
    print("Instantiating StableDiffusion model with jit_compile=False...")
    model = keras_cv.models.StableDiffusion(jit_compile=False)
    print("StableDiffusion model instantiated successfully.")
except Exception as e:
    print(f"Error during model instantiation: {e}")
    # 모델 인스턴스화 실패 시 여기서 중단
    raise

# 3. 텍스트 프롬프트 정의
prompt_1 = "A watercolor painting of a Golden Retriever at the beach"
prompt_2 = "A still life DSLR photo of a bowl of fruit"
prompt_3 = "The eiffel tower in the style of starry night"
prompt_4 = "An architectural sketch of a skyscraper"

interpolation_steps = 5

# 4. 텍스트 프롬프트 인코딩 시도
try:
    print("Encoding text prompts...")
    encoding_1 = tf.squeeze(model.encode_text(prompt_1))
    encoding_2 = tf.squeeze(model.encode_text(prompt_2))
    encoding_3 = tf.squeeze(model.encode_text(prompt_3))
    encoding_4 = tf.squeeze(model.encode_text(prompt_4))
    print("Text prompts encoded successfully.")

    # 인코딩된 텐서의 형태 출력
    print(f"Encoding shape (e.g., encoding_1): {encoding_1.shape}")

    # tf.linspace 사용법 수정 (예: encoding_1과 encoding_2 사이 보간)
    interpolated_encodings_1_to_2 = tf.linspace(encoding_1, encoding_2, interpolation_steps)
    print(f"Shape of interpolated encodings (1 to 2): {interpolated_encodings_1_to_2.shape}")

except Exception as e:
    print(f"Error during text encoding or linspace: {e}")
    # 오류 발생 시 상세 정보 출력
    import traceback
    traceback.print_exc()

!pip uninstall -y keras-cv tensorflow keras
!pip install tensorflow==2.15.0 keras==2.15.0 keras-cv==0.8.2

````

Once we've interpolated the encodings, we can generate images from each point.
Note that in order to maintain some stability between the resulting images we
keep the diffusion noise constant between images.

````python
import tensorflow as tf
import keras_cv
import keras
import matplotlib.pyplot as plt
import math

# --- Cell 1 에서 model 과 interpolated_encodings_1_to_2 가 정의되었다고 가정 ---
# model = keras_cv.models.StableDiffusion(jit_compile=False)
# encoding_1 = tf.squeeze(model.encode_text(prompt_1)) # Cell 1의 결과 중 첫 번째 인코딩

# Cell 2 시작 (단일 이미지 생성 테스트)

# 0. 사용할 단일 인코딩 (예: 보간된 결과 중 첫 번째 또는 원본 인코딩_1)
#    interpolated_encodings_1_to_2[0]의 shape은 (77, 768)이 됩니다.
#    generate_image는 (batch_size, 77, 768)을 기대하므로 expand_dims 필요
single_encoding_for_generation = tf.expand_dims(interpolated_encodings_1_to_2[0], axis=0) # Shape: (1, 77, 768)

seed = 12345

# 1. 단일 노이즈 생성
#    diffusion_noise는 generate_image에 전달될 때 (batch_size, height/8, width/8, 4) 형태를 기대합니다.
#    여기서는 batch_size=1 입니다.
single_diffusion_noise = tf.random.normal(
    (1, 512 // 8, 512 // 8, 4), # (1, 64, 64, 4)
    seed=seed
)

# 2. Negative prompt 설정
negative_prompt_for_single = "" # 단일 문자열

print(f"Generating 1 image...")
print(f"Shape of single_encoding_for_generation: {single_encoding_for_generation.shape}")
print(f"Shape of single_diffusion_noise: {single_diffusion_noise.shape}")

try:
    image_batch = model.generate_image(
        encoded_text=single_encoding_for_generation, # 배치 크기 1인 인코딩
        negative_prompt=negative_prompt_for_single,
        num_steps=25,
        diffusion_noise=single_diffusion_noise, # 배치 크기 1인 노이즈
    )
    print("Image generated successfully!")
    image = image_batch[0] # 배치 중 첫 번째 (이자 유일한) 이미지

    # 생성된 이미지 표시
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title("Single Generated Image")
    plt.axis("off")
    plt.show()

except Exception as e:
    print(f"Error during single image generation: {e}")
    import traceback
    traceback.print_exc()
````

````python
# 이전 셀에서 'image' 변수에 단일 이미지가 numpy 배열로 성공적으로 생성되었다고 가정합니다.
# 예: image = image_batch[0] 와 같이요.
# 만약 'image' 변수가 정의되지 않았다면, 이 코드 실행 전에 이미지 생성 셀을 다시 실행해야 합니다.

from PIL import Image # PIL.Image 모듈을 임포트합니다.
# from IPython.display import Image as IImage # GIF 표시를 위해 필요할 수 있습니다.

# export_as_gif 함수 정의 (이미 정의했다면 다시 정의할 필요는 없을 수 있습니다)
def export_as_gif(filename, pil_image_list, frames_per_second=10, rubber_band=False):
    """
    PIL Image 객체들의 리스트를 GIF 파일로 저장합니다.
    pil_image_list: PIL.Image 객체들의 리스트여야 합니다.
    """
    if not pil_image_list:
        print("저장할 이미지가 없습니다.")
        return

    # rubber_band 로직 수정 (이미지가 충분히 있을 때만 적용)
    if rubber_band and len(pil_image_list) >= 2: # 최소 2장 이상일 때 의미가 있음
        # 예: A, B, C -> A, B, C, B (마지막 제외하고 뒤집어서 추가)
        # 또는 A, B, C, D -> A, B, C, D, C, B
        to_append_reversed = pil_image_list[1:-1] # 처음과 마지막 제외 (또는 마지막만 제외 등)
        if to_append_reversed: # 추가할 프레임이 있다면
             pil_image_list += to_append_reversed[::-1]
        # 원래 코드: images += images[2:-1][::-1]
        # 이 부분은 이미지 개수에 따라 다르게 동작할 수 있으므로, 의도에 맞게 수정 필요 시 알려주세요.
        # 현재는 간단히 두 번째부터 마지막에서 두 번째 프레임까지를 뒤집어 추가하는 방식으로 가정합니다.
        # 더 간단하게는: if len(pil_image_list) > 1: pil_image_list += pil_image_list[:-1][::-1] (마지막 제외하고 전부 뒤집어 추가)

    pil_image_list[0].save(
        filename,
        save_all=True,
        append_images=pil_image_list[1:],
        duration=1000 // frames_per_second,
        loop=0,
    )
    print(f"GIF 파일이 '{filename}'으로 저장되었습니다.")


# --- 이미지 생성 및 GIF 변환 ---
try:
    # 1. 이전에 생성된 단일 이미지(numpy 배열)를 'image' 변수에서 가져옵니다.
    #    만약 'image' 변수가 없다면 NameError가 여기서 발생합니다.
    #    이 경우, 이미지 생성 코드를 먼저 실행해야 합니다.
    #    예를 들어, 이전 단계에서 image = image_batch[0] 와 같이 단일 이미지를 저장했습니다.

    # 'image' 변수가 정의되어 있다고 가정하고, 이를 리스트에 넣습니다.
    # 'image'는 numpy 배열이어야 합니다.
    numpy_image_list_for_gif = [image]

    # 2. Numpy 배열 리스트를 PIL Image 객체 리스트로 변환합니다.
    pil_images = [Image.fromarray(img_arr) for img_arr in numpy_image_list_for_gif]

    # 3. GIF로 저장합니다.
    export_as_gif(
        "doggo-and-fruit-single.gif", # 파일 이름을 다르게 해서 테스트해보세요.
        pil_images,
        frames_per_second=2, # 초당 프레임 수
        rubber_band=True, # rubber_band 효과를 주려면 이미지가 여러 장 필요합니다.
                           # 현재는 이미지가 한 장이므로 rubber_band 효과는 크게 나타나지 않습니다.
    )

    # 4. (선택 사항) 생성된 GIF를 노트북에 표시합니다.
    # from IPython.display import Image as IImage
    # display(IImage(filename="doggo-and-fruit-single.gif"))

except NameError:
    print("오류: 'image' 변수가 정의되지 않았습니다. 이미지 생성 셀을 먼저 실행하고,")
    print("생성된 단일 이미지가 'image'라는 변수에 저장되어 있는지 확인해주세요.")
except Exception as e:
    print(f"GIF 생성 중 오류 발생: {e}")
    import traceback
    traceback.print_exc()
````

The results may seem surprising. Generally, interpolating between prompts
produces coherent looking images, and often demonstrates a progressive concept
shift between the contents of the two prompts. This is indicative of a high
quality representation space, that closely mirrors the natural structure
of the visual world.

To best visualize this, we should do a much more fine-grained interpolation,
using hundreds of steps. In order to keep batch size small (so that we don't
OOM our GPU), this requires manually batching our interpolated
encodings.

이 블럭은 25분 정도 소요됩니다!
배치의 크기를 줄여서 테스트해보길 추천드릴게요!
노트북의 파일은 기존의 결과물로, 새로 테스트하는 결과물과 배치 크기에 따른 결과를 비교해보는 것도 좋은 실험이겠네요 😀

````python
import tensorflow as tf
import keras_cv # 이미 임포트 되어 있어야 함
import keras    # 이미 임포트 되어 있어야 함
from PIL import Image # GIF 생성을 위해 필요
# import matplotlib.pyplot as plt # 각 이미지 확인 시 필요할 수 있음
# import math # 필요시 사용

# --- 이전 셀에서 model, encoding_1, encoding_2 변수가 성공적으로 정의되었다고 가정 ---
# 예 (이런 코드들이 이전 셀에 있어야 합니다):
# print(f"TensorFlow version: {tf.__version__}")
# print(f"Keras version: {keras.__version__}")
# print(f"KerasCV version: {keras_cv.__version__}")
# keras.mixed_precision.set_global_policy("float32")
# model = keras_cv.models.StableDiffusion(jit_compile=False) # model 정의
# prompt_1 = "A watercolor painting of a Golden Retriever at the beach"
# prompt_2 = "A still life DSLR photo of a bowl of fruit"
# encoding_1 = tf.squeeze(model.encode_text(prompt_1)) # encoding_1 정의
# encoding_2 = tf.squeeze(model.encode_text(prompt_2)) # encoding_2 정의
# --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- --- ---

# export_as_gif 함수 정의 (이미 현재 세션에 정의되어 있다면 다시 정의할 필요 없음)
def export_as_gif(filename, pil_image_list, frames_per_second=10, rubber_band=False):
    if not pil_image_list:
        print("저장할 이미지가 없습니다.")
        return
    if rubber_band and len(pil_image_list) >= 2:
        to_append_reversed = pil_image_list[1:-1] # 처음과 마지막 프레임 제외
        if to_append_reversed:
             pil_image_list += to_append_reversed[::-1] # 뒤집어서 추가

    pil_image_list[0].save(
        filename,
        save_all=True,
        append_images=pil_image_list[1:],
        duration=1000 // frames_per_second,
        loop=0,
    )
    print(f"GIF 파일이 '{filename}'으로 저장되었습니다.")

# --- 150개 이미지 생성 및 GIF 변환 시작 (단일 이미지 생성 반복) ---
try:
    interpolation_steps = 6

    # 1. 텍스트 인코딩 보간 (tf.linspace 사용)
    print(f"Interpolating text encodings for {interpolation_steps} steps...")
    # encoding_1과 encoding_2가 현재 범위에 정의되어 있어야 합니다.
    interpolated_encodings = tf.linspace(encoding_1, encoding_2, interpolation_steps) # Shape: (10, 77, 768)
    print(f"Shape of interpolated_encodings: {interpolated_encodings.shape}")

    all_generated_pil_images = []
    base_seed = 12345 # 노이즈 생성을 위한 기본 시드
    negative_prompt_for_iterations = "" # 모든 반복에 동일한 negative_prompt 사용

    print(f"\nStarting iterative image generation for {interpolation_steps} images...")
    for i in range(interpolation_steps):
        print(f"  Generating image {i+1}/{interpolation_steps}...")

        current_single_encoding = tf.expand_dims(interpolated_encodings[i], axis=0)
        current_single_noise = tf.random.normal(
            (1, 512 // 8, 512 // 8, 4),
            seed=base_seed + i
        )

        # print(f"    Shape of current_single_encoding: {current_single_encoding.shape}")
        # print(f"    Shape of current_single_noise: {current_single_noise.shape}")

        # <<< 들여쓰기 수정: try 블록을 for 루프 내의 다른 코드와 동일한 수준으로 맞춤 >>>
        try:
            generated_image_batch_np = model.generate_image(
                encoded_text=current_single_encoding,
                num_steps=25,
                diffusion_noise=current_single_noise,
                negative_prompt=negative_prompt_for_iterations,
            )
            single_image_np = generated_image_batch_np[0] # .numpy() 제거됨

            all_generated_pil_images.append(Image.fromarray(single_image_np))
            print(f"    Image {i+1} generated successfully.")

        except Exception as e_inner:
            print(f"    Error generating image {i+1}: {e_inner}")
            import traceback
            traceback.print_exc()
            pass

    print("\nAll image generation iterations complete.")

    if all_generated_pil_images:
        export_as_gif("doggo-and-fruit-150.gif", all_generated_pil_images, frames_per_second=10, rubber_band=True)
    else:
        print("No images were successfully generated to create a GIF.")

except NameError as ne:
    print(f"NameError 발생: {ne}")
    print("오류가 발생한 변수(예: 'model', 'encoding_1', 'encoding_2')가 현재 셀 실행 전에 정의되었는지 확인해주세요.")
    print("필요하다면 해당 변수들을 정의하는 이전 셀을 다시 실행해주세요.")
    import traceback
    traceback.print_exc()
except Exception as e:
    print(f"An error occurred: {e}")
    import traceback
    traceback.print_exc()
````

````python
import tensorflow as ops
import numpy as np # Make sure numpy is imported
import matplotlib.pyplot as plt # Make sure matplotlib.pyplot is imported

# ... (your prompts and other initializations) ...
prompt_1 = "A watercolor painting of a Golden Retriever at the beach"
prompt_2 = "A still life DSLR photo of a bowl of fruit"
prompt_3 = "The eiffel tower in the style of starry night"
prompt_4 = "An architectural sketch of a skyscraper"

interpolation_steps = 6
batch_size = 3
batches = (interpolation_steps**2) // batch_size

# Presumed: model is already defined/loaded here

encoding_1 = ops.squeeze(model.encode_text(prompt_1))
encoding_2 = ops.squeeze(model.encode_text(prompt_2))
encoding_3 = ops.squeeze(model.encode_text(prompt_3))
encoding_4 = ops.squeeze(model.encode_text(prompt_4))

interpolated_encodings = ops.linspace(
    ops.linspace(encoding_1, encoding_2, interpolation_steps),
    ops.linspace(encoding_3, encoding_4, interpolation_steps),
    interpolation_steps,
)
interpolated_encodings = ops.reshape(
    interpolated_encodings, (interpolation_steps**2, 77, 768) # Assuming 77, 768 are token length and embedding dim
)
batched_encodings = ops.split(interpolated_encodings, batches)

# Define noise
# IMPORTANT: Adjust these latent dimensions according to your model's requirements.
# These are example values.
latent_height = 64
latent_width = 64
latent_channels = 4 # Common for latent diffusion models

# Option A: Generate one noise tensor and reuse it for all batches.
# This helps in seeing the effect of interpolated encodings more clearly,
# as the starting noise is consistent.
# The model.generate_image function would need to correctly handle this single noise
# for all items in the batch (e.g. by tiling it or if it expects one noise for the batch).
# Ensure the batch_size here matches what generate_image expects for its noise input.
noise = ops.random.normal(shape=(batch_size, latent_height, latent_width, latent_channels))

# Option B: If you want different noise for each image generated across all batches (more complex setup)
# You might generate noise for ALL images upfront and then slice it in the loop:
# total_images = interpolation_steps**2
# all_noise = ops.random.normal(shape=(total_images, latent_height, latent_width, latent_channels))
# Then inside the loop, you'd use a slice of all_noise for the current batch.
# For now, Option A is simpler if `noise` is meant to be defined once.

images = []
for i in range(batches): # Corrected loop variable from 'batch' to 'i' to avoid conflict if 'batches' is a number
    current_batch_encodings = batched_encodings[i]
    # Determine the actual size of the current batch, as the last batch might be smaller
    current_batch_size = ops.shape(current_batch_encodings)[0].numpy()

    # If using Option A noise (fixed for all batches):
    # You might need to ensure your model.generate_image can take a noise tensor
    # that might be for a 'full' batch_size even if current_batch_size is smaller,
    # or you might slice the noise:
    # current_noise = noise[:current_batch_size]
    # However, many models expect the noise to match the batch size precisely.
    # For simplicity with Option A, assuming model.generate_image handles a fixed `noise`
    # or that `batch_size` passed to generate_image aligns with `noise`'s first dim.
    # If `noise` is generated for a fixed `batch_size` (as in Option A),
    # and `model.generate_image` processes `current_batch_size` items,
    # you must ensure compatibility. It might be safer to generate noise
    # matching `current_batch_size` if the model is strict.

    # Let's assume the model.generate_image will use the passed `batch_size` argument
    # to determine how to use the `noise` or the `current_batch_encodings`.
    # Passing the globally defined 'noise' (from Option A) directly:
    images.append(
        model.generate_image(
            current_batch_encodings,
            batch_size=current_batch_size, # Pass the actual size of the current batch
            diffusion_noise=noise, # Using the 'noise' defined before the loop
        )
    )

# ... (plot_grid function and the rest of your code) ...
def plot_grid(
    images_to_plot, # Renamed to avoid conflict with outer 'images' list
    path,
    grid_size,
    scale=2,
):
    fig, axs = plt.subplots(
    grid_size, grid_size, figsize=(grid_size * scale, grid_size * scale)
    )
    fig.tight_layout()
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.margins(x=0, y=0)
    # plt.axis("off") # Should be on ax, not plt overall for subplots
    for ax_idx, ax in enumerate(axs.flat):
        ax.axis("off")
        if ax_idx < len(images_to_plot):
            # Ensure images are in the correct format (e.g., uint8) and range [0, 255]
            img_data = images_to_plot[ax_idx]
            if isinstance(img_data, ops.Tensor):
                img_data = img_data.numpy() # Convert to numpy if it's a tensor

            # Basic normalization if images are float in [0,1] or other ranges
            if img_data.max() <= 1.0 and img_data.min() >=0.0 :
                 img_data = (img_data * 255)

            img_data = img_data.astype("uint8")
            ax.imshow(img_data)
    plt.savefig(
        fname=path,
        pad_inches=0,
        bbox_inches="tight",
        transparent=False,
        dpi=60,
    )

images_np = np.concatenate([img.numpy() if hasattr(img, 'numpy') else img for img in images], axis=0)
plot_grid(images_np, "4-way-interpolation.jpg", interpolation_steps)
````

## A circular walk through the diffusion noise space for a single prompt

Our final experiment is to stick to one prompt and explore the variety of images
that the diffusion model can produce from that prompt. We do this by controlling
the noise that is used to seed the diffusion process.

We create two noise components, `x` and `y`, and do a walk from 0 to 2π, summing
the cosine of our `x` component and the sin of our `y` component to produce noise.
Using this approach, the end of our walk arrives at the same noise inputs where
we began our walk, so we get a "loopable" result!

````python
prompt = "An oil paintings of cows in a field next to a windmill in Holland"
encoding = tf.squeeze(model.encode_text(prompt))
walk_steps = 150
batch_size = 3
batches = walk_steps // batch_size

walk_noise_x = tf.random.normal(noise.shape, dtype=tf.float64)
walk_noise_y = tf.random.normal(noise.shape, dtype=tf.float64)

walk_scale_x = tf.cos(tf.linspace(0, 2, walk_steps) * math.pi)
walk_scale_y = tf.sin(tf.linspace(0, 2, walk_steps) * math.pi)
noise_x = tf.tensordot(walk_scale_x, walk_noise_x, axes=0)
noise_y = tf.tensordot(walk_scale_y, walk_noise_y, axes=0)
noise = tf.add(noise_x, noise_y)
batched_noise = tf.split(noise, batches)
````
